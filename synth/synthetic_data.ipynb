# Синтетические данные о продажи недвижимости

# Цель проекта

Создание синтетического датасета, который репрезентирует закрытые данные о продаже недвижимости в Новосибирске, с добавлением геопризнаков. Синтетический датасет (необходимый размер: 9000 квартир) будет использован в рамках курса "Геопространственный анализ данных" https://ods.ai/tracks/geoanalytics-course-spring24. 

*Описание данных*

Данные состоят из файлов, полученных из разных источников:

- `csv-8606.csv.bz2` - цены на квартиры в Новосибирске, 2019 год, сырые данные;
- `export-reestrmkd-54-20240401.zip` - реестр многоквартирных домов Новосибирской области, с сайта https://витрина.фрт.рф/ получены 25.04.2024
- `tags.parquet` - выгруженные данные из OpenStreetMap от 25.04.2024

# План решения задачи 

1) Подготовить данные о продаже недвижимости в Новосибирске за 2019 год:

 - Извлечение ключевых признаков из данных: площадь кухни, количество комнат, тип объекта, общая площадь, тип дома, географические координаты, стоимость, информация о первом и последнем этажах;
 - Удаление дубликатов, обработка пропущенных значений, анализ данных и устранение выбросов и аномалий.
 
 
2) Провести геокодирование данных о многоквартирных домах:

- Получение данных о многоквартирных домах Новосибирской области из открытых источников;
- Геокодирование данных с помощью DaData.

3) Провести геопространственную индексацию данных о продаже квартир и многоквартирных домов:

- Группировка данных по гексагонам;
- Расчет математического ожидания продажи квартиры для каждого дома;

4) Провести гененерацию синтетических квартир;


5) Добавить пространственные признаки с использованием данных c OpenStreetMap:

- Создать пространственные признаки для гексагонов 9 и 11 уровней Новосибирска.
- Построить модель предсказания цен на квартиры на исходных данных с добавлением пространственных признаков.
- Оценить качество модели 

6) Добавить предсказанные цены с учетом динамики цен и нормального шума в синтетический датасет.


7) Построить модель для предсказания цен на квартиры и оценить ее качество.

### Формирование данных

Сформируем данные из "сырых" данных по признакам.

```python
df = pd.DataFrame(dataraw['Доп.параметры'])
df_normalized = json_normalize(df['Доп.параметры'])
df = pd.concat([df.drop(['Доп.параметры'], axis=1), json_normalize(df['Доп.параметры'])], axis=1)

dd = pd.concat([dataraw['lat'], dataraw['lng'], dataraw['Цена']], axis=1)

flats = pd.concat([df, dd], axis=1).drop(columns=['Тип объявления', 'Жилая площадь',\
                                                 'Срок сдачи'], axis=1)
flats.head()
```

заменим пропуски с площадью кухни, умножив площадь на коэффиент, характерный для группы по количеству комнат.

```python
charact_koef = dict(zip(data_grouped['rooms'], data_grouped['part_of_kitchen']))
charact_koef
```

```python
def square_kitchen(row):
    if pd.isnull(row['square_kitchen']):
        num_rooms = row['rooms']
        if num_rooms in charact_koef:
            koef = charact_koef[num_rooms]  
            square_kitchen = row['square'] * koef 
            return square_kitchen
    return row['square_kitchen'] 

flats['square_kitchen'] = flats.apply(square_kitchen, axis=1)
```

Совпадающие признаки дома и квартиры берем из дома.

Функция набора фильтров для запроса:

```python
def filter_by_query(df, query):
    if query:
        chunks = []
        for k, v in query.items():
            if isinstance(v, str):
                chunk = f'{k}=="{v}"'
                
            else:
                chunk = f'{k}=={v}'
            chunks.append(chunk)
                
        return df.query(" & ".join(chunks))
    else:
        return df
```

Сэмплирование:

```python
flats_synth = buildings[['lat', 'lng', 'type_of_house', 'object_type']].copy()
records_list = []

columns = ['floor_first',  'floor_last',  'square_range',  'square_kitchen_range', 'rooms']


for i in tqdm(range(len(buildings)), desc='Progress'):
    record = dict()
    query_house = {'type_of_house': flats_synth.iloc[i]['type_of_house'], 'object_type': flats_synth.iloc[i]['object_type']}
    query = dict(query_house)
    random.shuffle(columns) 
    
    for column in columns:
        sample = filter_by_query(flats, query).sample(1)
        if column == 'square_range':
            value_column = sample[column].values[0]
            query[column] = value_column
            record['square'] = sample['square'].values[0]
            #print(query)
            
        elif column == 'square_kitchen_range':
            value_column = sample[column].values[0]
            query[column] = value_column
            record['square_kitchen'] = sample['square_kitchen'].values[0]
            #print(query)
            
        else:
            value_column = sample[column].values[0]
            query[column] = value_column
            record[column] = value_column
            #print(query)
            
    #print(record)        
    records_list.append(record)


data_records = pd.DataFrame(records_list)
flats_synth = pd.concat([flats_synth, data_records], axis=1)
```

```python
plt.scatter(target_test, mean_test, color="tab:blue", alpha = 0.2)
plt.axvline(x = mean_train, color = "tab:red", linestyle='--')

plt.annotate(f'RMSE: {rmse:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12, ha='left', va='top', color="tab:red")
plt.annotate(f'MAE: {mae:.2f}', xy=(0.05, 0.85), xycoords='axes fraction', fontsize=12, ha='left', va='top', color="tab:red")

plt.title('scatter plot: prediction by average')
plt.xlabel('actual values')
plt.ylabel('predicted values')

plt.show()
```
